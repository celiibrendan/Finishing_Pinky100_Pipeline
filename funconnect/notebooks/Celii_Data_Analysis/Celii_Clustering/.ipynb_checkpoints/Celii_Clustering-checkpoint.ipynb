# Initial Generation of Axon Feature Vectors
- `relevant_axon_trace_synapses`: Properly restricted Synapse Table (labeled segments are the postsyns)
- `synapse_ids`: np.array of synapse ids
- `presyns`: np.array of presynaptic ids (axons)
- `postsyns`: np.array of postsynaptic ids (neurons)
- `postsyn_to_synapses`: dict of neuron ids to synapses
- `synapse_label_dict`: dict of synapse ids to their compartment labels
- `axon_to_synapse`: dict mapping each axon to their synapses
- `descriptions`: dict mapping feature in axon feature vectors to synapse contacting compartment description (Apicals, Basals, Obliques, Somas, Axons)
- `axon_to_vector`: dict mapping of each axon to their feature vectors (which consist of the number of synapses in each compartment specified by description)
- `threshold_axon_to_vector`: dict mapping of each thresholded axon to their feature vectors (this stores only axons with more than three synapses of the specified compartments)
- `axon_compartment_vector`: np.ndarray of all of the axon feature vectors
- `threshold_axon_compartment_vector`: np.ndarray of all of the thresholded axon feature vectors
- `fingerprints_array`: np.ndarray of all of the axon feature vectors with each vector being normalized
- `threshold_fingerprints_array`: np.ndarray of all of the thresholded axon feature vectors with each vector being normalized
- `neuron_to_axon`: dict mapping of neuron ids to the axons that synapse onto them
- `neuron_to_vector`: dict mapping of neuron ids to the vectorized sum of the axon feature vectors that synapse onto them at any point
- `cmaps_qualitative`: list of qualitative cmaps for use in matplotlib plots

# Christos Code

```python
#gets unique list of axons
axon_segments = (ta3.Neurite() & 'neurite_type="axon"').proj(presyn='segment_id')

#list of 
relevant_axon_trace_synapses = ta3.Synapse & axon_segments & ta3.ProofreadLabel.proj(postsyn='segment_id')

synapse_ids, presyns, postsyns = relevant_axon_trace_synapses.fetch('synapse_id', 'presyn', 'postsyn')
presyns_counter, postsyns_counter = Counter(presyns), Counter(postsyns)
synapse_labels = ta3.SynapseCompartmentLabel().fetch('synapse_id', 'postsynaptic_label')

postsyns_to_synapses = dict()
for postsyn in tqdm(postsyns_counter):
    postsyns_to_synapses[postsyn] = synapse_ids[np.where(postsyns==postsyn)[0]]
    
synapse_label_dict = dict()
for synapse_id, label in tqdm(zip(*synapse_labels)):
    synapse_label_dict[synapse_id] = label

axon_to_synapse = dict()
for synapse, presyn in tqdm(zip(synapse_ids, presyns)):
    if presyn in axon_to_synapse:
        axon_to_synapse[presyn].append(synapse)
    else:
        axon_to_synapse[presyn] = list()
        axon_to_synapse[presyn].append(synapse)

descriptions = {0:'Apicals', 1:'Basals', 2:'Obliques', 3:'Somas', 4:'Axons'}
axon_to_vector = dict()
threshold_axon_to_vector = dict()
for presyn in tqdm(axon_to_synapse):
    label_list = list()
    for synapse in axon_to_synapse[presyn]:
        label_list.append(synapse_label_dict[synapse])
    label_counter = Counter(label_list)
    apicals   = label_counter[2]
    basals    = label_counter[3]
    obliques  = label_counter[4]
    somas     = label_counter[5]
    axons     = label_counter[6] + label_counter[7]
    synapse_total = sum((apicals, basals, obliques, somas, axons))
    if synapse_total > 0:
        axon_to_vector[presyn] = np.array((apicals, basals, obliques, somas, axons))
        if synapse_total > 3:
            threshold_axon_to_vector[presyn] = np.array((apicals, basals, obliques, somas, axons))
axon_compartment_vector = np.array(list(axon_to_vector.values()))
threshold_axon_compartment_vector = np.array(list(threshold_axon_to_vector.values()))

fingerprints_array = np.zeros(axon_compartment_vector.shape)
for i, axon_vector in tqdm(enumerate(axon_compartment_vector)):
    fingerprints_array[i] = axon_vector / axon_vector.sum()

threshold_fingerprints_array = np.zeros(threshold_axon_compartment_vector.shape)
for i, axon_vector in tqdm(enumerate(threshold_axon_compartment_vector)):
    threshold_fingerprints_array[i] = axon_vector / axon_vector.sum()

neuron_to_axon = dict()
for presyn, postsyn in tqdm(zip(presyns, postsyns)):
    if presyn in axon_to_vector:
        if postsyn in neuron_to_axon:
            neuron_to_axon[postsyn].append(presyn)
        else:
            neuron_to_axon[postsyn] = list()
            neuron_to_axon[postsyn].append(presyn)
    
neuron_to_vector = dict()
for neuron in tqdm(neuron_to_axon):
    neuron_to_vector[neuron] = np.array([axon_to_vector[axon] for axon in neuron_to_axon[neuron]], np.uint32).sum(axis=0)

cmaps_qualitative = ['Pastel1', 'Pastel2', 'Paired', 'Accent', 'Dark2', 'Set1', 'Set2', 'Set3', 'tab10', 'tab20', 'tab20b', 'tab20c']
```

```python
# kind = { "scatter" | "reg" | "resid" | "kde" | "hex" }
def plot_joint_marginals(data, kind='scatter', color="#4CB391", descriptions=None, **kwargs):
    for i, j in np.unique([sorted((i, j)) for i in range(data.shape[1]) for j in range(data.shape[1]) if i is not j], axis=0):
        column_a = data[:, i]
        column_b = data[:, j]
        removed_double_zeros_data = np.array([[a, b] for a, b in zip(column_a, column_b) if (a != 0) or (b != 0)]).T
        fig = sns.jointplot(removed_double_zeros_data[0], removed_double_zeros_data[1], kind=kind, color=color, **kwargs)
#         fig = sns.jointplot(data[:, i], data[:, j], kind=kind, color=color, **kwargs)
        fig.fig.suptitle(f"Jointplot: {i} vs {j}, {descriptions[i]} vs {descriptions[j]}", fontsize = 20, y = 1.05)
        if descriptions is not None:
            fig.set_axis_labels(xlabel=descriptions[i], ylabel=descriptions[j])
        plt.show()
```

## Cluster Stability and PCA

```python
input_axons = threshold_axon_compartment_vector #axon_compartment_vector
clusterer_input = input_axons
random_data = np.random.randint(24, size=threshold_axon_compartment_vector.shape)
random_data = abs(np.random.normal(scale=5, size=threshold_axon_compartment_vector.shape)).astype(np.uint8)
input_axons = random_data
# clusterer_input = random_data
random_data.max()
```

```python
clusterer = KMeans(n_clusters=8, n_init=10, random_state=0)
clusterer.fit(clusterer_input)
clusterer.labels_
```

```python
scores_per_cluster = np.zeros((input_axons.shape[0], clusterer.n_clusters))
for i, feature_vector in enumerate(input_axons):
    for j, center in enumerate(clusterer.cluster_centers_):
        scores_per_cluster[i, j] = np.linalg.norm(center-feature_vector)
scores_per_cluster = np.exp(-scores_per_cluster)
# scores_per_cluster = scores_per_cluster
groupings = clusterer.labels_#np.argmin(scores_per_cluster, axis=1)
groupings_dict = dict()
for i in np.unique(groupings):
    groupings_dict[i] = list()
for i, cluster in enumerate(groupings):
    groupings_dict[cluster].append(i)
mean_scores_per_group = np.zeros((clusterer.n_clusters, clusterer.n_clusters))
for i, axons in groupings_dict.items():
    group_scores = scores_per_cluster[np.array(axons)]
    for j, row in enumerate(group_scores):
        group_scores[j] /= row.sum()
    for j, column in enumerate(group_scores.T):
        mean_scores_per_group[i, j] = column.mean()
for i, row in enumerate(mean_scores_per_group):
    print("Expected Cluster:", i, "-> index of max:", np.argmax(row))
sns.heatmap(mean_scores_per_group)
```

```python
pca = PCA(n_components=4, random_state=0)
principalComponents = pca.fit_transform(input_axons).T
plt.scatter(principalComponents[0], principalComponents[1], c=clusterer.labels_, cmap=plt.get_cmap(cmaps_qualitative[5]), alpha=0.5)
```

```python
def threshold_feature_vectors(feature_vectors, lower_threshold=0, upper_threshold=NotImplemented):
    thresholded_features = dict()
    for key_id, feature_vector in feature_vectors.items():
        if feature_vector.sum() > lower_threshold:
            thresholded_features[key_id] = feature_vector
    return thresholded_features

def normalize_feature_vectors(feature_vectors):
    normalized_features = dict()
    for key_id, feature_vector in feature_vectors.items():
        normalized_features[key_id] = feature_vector / feature_vector.sum()
    return normalized_features
```

```python
# Extremely fast implementation, Linear time, 2 million iterations through the arrays per second.
# Should add another check just like what I have up above to ensure previous thresholding is upheld. This reduces performance
# to 1.5 million iterations per second.
def group_by_unique_values(array1, array2, enable_tqdm=False):
    array1_to_values_in_array2 = dict()
    for element_in_array1, element_in_array2 in tqdm(zip(array1, array2, disable=not enable_tqdm)):
        if element_in_array1 in array1_to_values_in_array2:
            array1_to_values_in_array2[element_in_array2].append(element_in_array1)
        else:
            array1_to_values_in_array2[element_in_array2] = list()
            array1_to_values_in_array2[element_in_array2].append(element_in_array1)
    return array1_to_values_in_array2
```
