```python
#loop that will insert the errors into the Synapse exclude
#1) get the synapses with Error label
#2) for each synapse, pull down the data from Synapse table and then write it to the SynapseExclude tabel (ignore duplicates)
synapse_ids_error = (ta3p100.SynapseCompartmentLabel() & "segmentation=2" & "postsynaptic_label=10").fetch("synapse_id")
print(len(synapse_ids_error))

from tqdm import tqdm

total_dict = []
for err in tqdm(synapse_ids_error):
    new_err_syn = (ta3p100.Synapse() & "synapse_id="+str(err) & "segmentation=2").fetch()
    #print(new_err_syn[0])
    #make the insert
    ta3p100.SynapseExclude().insert1(new_err_syn[0],skip_duplicates=True)
```
